**Data Cleaning**

```python
df['REPAIR_DATE'] = pd.to_datetime(df['REPAIR_DATE'], errors='coerce')
```

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip().str.upper()
```

```python
cat_fill_cols = [
    'CAUSAL_PART_NM',
    'PLANT',
    'STATE',
    'REPAIR_DLR_POSTAL_CD',
    'VEH_TEST_GRP',
    'OPTN_FAMLY_CERTIFICATION',
    'OPTF_FAMLY_EMISSIOF_SYSTEM',
    'ENGINE_SOURCE_PLANT',
    'ENGINE_TRACE_NBR',
    'TRANSMISSION_TRACE_NBR',
    'LINE_SERIES'
]

df[cat_fill_cols] = df[cat_fill_cols].fillna("UNKNOWN")
```

```python
num_fill_cols = [
    'TOTALCOST',
    'TRANSMISSION_SOURCE_PLANT',
    'LAST_KNOWN_DELVRY_TYPE_CD'
]

for col in num_fill_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.drop(columns=['CAMPAIGN_NBR'], inplace=True)
```

```python
df = df[df['KM'] >= 0]
df = df[df['TOTALCOST'] >= 0]
df = df[df['REPORTING_COST'] >= 0]
df = df[df['LBRCOST'] >= 0]
```

```python
Q1 = df['TOTALCOST'].quantile(0.25)
Q3 = df['TOTALCOST'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['TOTALCOST'] >= lower_bound) &
        (df['TOTALCOST'] <= upper_bound)]
```

**Component & Failure Tags**

```python
def generate_tags(text):
    tags = []
    text = str(text)

    if "STEERING WHEEL" in text:
        tags.append("STEERING_WHEEL")
    if "HARNESS" in text:
        tags.append("WIRING_HARNESS")
    if "MODULE" in text:
        tags.append("CONTROL_MODULE")
    if "COMING APART" in text or "FINISH COMING OFF" in text:
        tags.append("MATERIAL_DEGRADATION")
    if "INOPERABLE" in text or "NOT WORKING" in text:
        tags.append("FUNCTIONAL_FAILURE")
    if "MESSAGE" in text or "DTC" in text:
        tags.append("ELECTRONIC_FAULT")

    return ",".join(tags)

df['FAILURE_TAGS'] = df['CUSTOMER_VERBATIM'].apply(generate_tags)
```

**Early vs Late Failure**

```python
df['FAILURE_STAGE'] = np.where(
    df['KM'] <= 10000,
    'EARLY_LIFE_FAILURE',
    'MID_LIFE_FAILURE'
)
```

**High Cost Flag**

```python
df['HIGH_COST_FLAG'] = np.where(
    df['TOTALCOST'] > df['TOTALCOST'].median(),
    'HIGH_COST',
    'NORMAL_COST'
)
```
